### 1.1 Définition d'une matrice

1. Définir une matrice 

Une matrice de dimension (ou d'ordre or de taille) n×pn\times pn×p est un tableau de nombres réels (appelés coefficients ou termes) comportant nnn lignes et ppp colonnes.

2. Quelle est sa notation? 

Les éléments sont nommés en utilisant deux indices,le premier est l’indice de ligne, lesecond est l’indice de colonne. 

3. Définir une matrice colonne et une matrice ligne (faire le lien avec les vecteurs)

    Une matrice qui a une seule ligne est appelée matrice ligne. 

    Une matrice qui a une seule colonne est appelée matrice colonne

#### Matrices particulières

1. Qu'est-ce qu'une matrice carrée?

C'est une matrice qui contient autant de ligne que de colonne
2. Qu'est-ce qu'une matrice triangulaire supérieure?

On appelle matrice triangulaire supérieure (respectivement inférieure) une matrice carrée dont tous les termes " au-dessous " (respectivement " au-dessus ") de la diagonale principale sont nuls. 

3. Qu'est-ce qu'une matrice triangulaire inférieure?



4. Qu'est-ce qu'une matrice diagonale?

Une matrice diagonale une matrice carrée dont les termes situés hors de la diagonale principale sont tous nuls. 

5. Qu'est-ce qu'une matrice identité?

La matrice identité de dimension n*n, notée In, start subscript, n, end subscript, est une matrice carrée de n lignes et n colonnes. Tous les éléments de sa diagonale sont égaux à 1 et tous les autres éléments sont égaux à 0. 



### 1.2 Matrice d'une application linéaire de E dans F

1. Définir une matrice d'une application linéaire

Soient :

    E et F deux espaces vectoriels sur un corps commutatif K, de dimensions respectives n et m ;
    B = (e1, … , en) une base de E, C une base de F ;
    φ une application de E dans F.

Alors2 :

    l'application φ est linéaire si et seulement s'il existe une matrice A de Mm,n(K) telle que
    pour tout vecteur x de E, la colonne des coordonnées dans C de φ(x) soit le produit à gauche par A de la colonne de coordonnées de x dans B ;
    une telle matrice A est alors unique : ses n colonnes sont les coordonnées dans C des n vecteurs φ(e1), … , φ(en).

    Cette matrice A est appelée la matrice de φ dans le couple de bases (B, C) et notée matB, C(φ), ou parfois MCB(φ).

    Plus formellement, matB, C(φ) est caractérisée par :
    ∀ x ∈ E mat C ⁡ ( φ ( x ) ) = mat B , C ⁡ ( φ ) × mat B ⁡ ( x ) {\displaystyle \forall x\in E\quad \operatorname {mat} _{C}(\varphi (x))=\operatorname {mat} _{B,C}(\varphi )\times \operatorname {mat} _{B}(x)} {\displaystyle \forall x\in E\quad \operatorname {mat} _{C}(\varphi (x))=\operatorname {mat} _{B,C}(\varphi )\times \operatorname {mat} _{B}(x)}.
