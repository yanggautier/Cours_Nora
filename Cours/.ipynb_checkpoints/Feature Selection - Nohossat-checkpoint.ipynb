# Feature Selection

Dans ce notebook, nous allons tester différentes **méthodes de sélection de variables**. En apprentissage supervisé,cette étape cruciale permet d'améliorer les performances des modèles prédictifs, en excluant les variables non significatives ou redondantes.

Ces méthodes se déclinent en 3 familles :
- **Filter** methods
- **Wrapped** methods
- **Embedded** methods 

Nous allons étudier les méthodes les plus largement utilisées en ML. Nous allons appliquer ces méthodes au dataset **Breast Cancer Winconsin** puis comparer les performances des modèles obtenues.

Pour rappel, ce dataset contient les informations suivantes (*source Kaggle*):

1. ID number
2. Diagnosis (M = malignant, B = benign)
3. 32 Ten real-valued features are computed for each cell nucleus:

       a) radius (mean of distances from center to points on the perimeter)
       b) texture (standard deviation of gray-scale values)
       c) perimeter
       d) area
       e) smoothness (local variation in radius lengths)
       f) compactness (perimeter^2 / area - 1.0)
       g) concavity (severity of concave portions of the contour)
       h) concave points (number of concave portions of the contour)
       i) symmetry
       j) fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image, resulting in 30 features. For instance, field 3 is Mean Radius, field 13 is Radius SE, field 23 is Worst Radius.

## 3. Feature Selection

### 3.1 Filter methods

1. Expliquer le principe de ces méthodes. Donner des exemples de méthodes



Filter methods are generally used as a **data preprocessing step**. The selection of features is **independent of any machine learning algorithm**. Features give rank on the basis of statistical scores which tend to determine the features' correlation with the outcome variable.

Filter methods select features from a dataset **independently for any machine learning algorithm**. These methods rely only on the characteristics of these variables, so features are filtered out of the data before learning begins.
These methods are powerful and simple and help to quickly remove features— and they are generally the first step in any feature selection pipeline.  

advantages:
- ML model agnostic
- computationally fast

Filter methods are very good for eliminating irrelevant, redundant, constant, duplicated, and correlated features.  

**Univariate filter methods** : rank features. downside : may pick up redundant features , ex: VarianceThreshold (Constant Features), Duplicated Features, SelectKBest, 
Univariate ROC-AUC /RMSE

**Multivariate filter methods** : rank features but take into consideration correlation (Pearson, Spearman rank correlation, Kendall rank correlation) : Chi-squared Score, ANOVA Univariate Test, , Mutual Information

### 3.2 Wrapper methods

1. Expliquer le principe de ces méthodes. Donner des exemples de méthodes

There is another category of feature selection methods that select the most optimal features for the specified algorithm. Such methods are called wrapper methods.

Wrapper methods are based on greedy search algorithms as they evaluate all possible combinations of the features and select the combination that produces the best result for a specific machine learning algorithm. A downside to this approach is that testing all possible combinations of the features can be computationally very expensive, particularly if the feature set is very large.

advantages : 
- best fit for a specified algorithm

3 types of wrapper methods: 

- Step forward feature selection : SequentialFeatureSelector forward
- Step backwards feature selection : SequentialFeatureSelector backward
- Exhaustive feature selection : ExhaustiveFeatureSelector

As a rule of thumb, if the dataset is small, exhaustive feature selection method should be the choice, however, in case of large datasets, step forward or step backward feature selection methods should be preferred.

### 3.2 Embedded methods

1. Expliquer le principe de ces méthodes. Donner des exemples de méthodes

>But what if we could include the feature selection process in ML model training itself? That could lead us to even better features for that model, in a shorter amount of time. This is where embedded methods come into play. In other words, they perform feature selection during the model training, which is why we call them embedded methods.
A learning algorithm takes advantage of its own variable selection process and performs feature selection and classification/regression at the same time.

Advantages : 
- They take into consideration the interaction of features like wrapper methods do.
- They are faster like filter methods.
- They are more accurate than filter methods.
- They find the feature subset for the algorithm being trained.
- They are much less prone to overfitting.

2 methods : 
- regularization
- tree-based methods

## 4. Modeling et Validation

Appliquer la régression Logistique et Evaluer les performances des modèles (Accuracy, Recall, Precision, ROC, AUC) selon les scenarii suivants :

1. Sur l'ensemble des features (sans selection)

```python
def run_model(model, X_train, X_test, y_train, y_test, results, feat_select_type, features_selection = None, reg_lasso= None):
    
    # get pertinent features
    if features_selection:
        X_train = features_selection.transform(X_train)
        X_test = features_selection.transform(X_test)
        
    # fit model
    if reg_lasso:
        clf = model(penalty='l1', solver='liblinear', random_state=0).fit(X_train, y_train)
        model_name = f'{type(clf).__name__} - Lasso'
    else :
        clf = model(random_state=0).fit(X_train, y_train)
        model_name = type(clf).__name__
    
    # predictions on train and test
    y_pred = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    
    # get report for train
    names = y.astype('category').cat.categories.values
    report_train = classification_report(y_train, y_pred, target_names = names, output_dict = True)
    
    # get report for test
    report_test = classification_report(y_test, y_pred_test, target_names = names, output_dict = True)
    
    # Get AUC for test
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_test)
    auc_res = auc(fpr, tpr)
    
    # save results
    result_train = [f'{feat_select_type} - train', 
              model_name, 
              report_train['accuracy'], 
              report_train['weighted avg']['f1-score'], 
              report_train['weighted avg']['precision'], 
              report_train['weighted avg']['recall'],
              None]
    
    result_test = [f'{feat_select_type} - test', 
              model_name, 
              report_test['accuracy'], 
              report_test['weighted avg']['f1-score'], 
              report_test['weighted avg']['precision'], 
              report_test['weighted avg']['recall'],
              auc_res]
    
    results.append(result_train)
    results.append(result_test)
    
    # Compute ROC curve
    fig, ax = plt.subplots()
    plot_roc_curve(clf, X_test, y_test, name=f'ROC - {feat_select_type}',
                         alpha=0.3, lw=2, ax=ax)
    ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
        label='Chance', alpha=.8)
    
    return False
    
```

```python
results = []
```

```python
run_model(LogisticRegression, X_train, X_test, y_train, y_test, results, 'no preprocessing', None, None)
```

2. Sur les subsets de features définis aux différentes étapes

```python
for name, selector in feature_selection_fcts.items():
    run_model(LogisticRegression, X_train, X_test, y_train, y_test, results, name, selector, None)
```
