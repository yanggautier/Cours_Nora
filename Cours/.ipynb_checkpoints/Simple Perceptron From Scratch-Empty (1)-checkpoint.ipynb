# Simple Perceptron From Scratch

L'objectif de ce notebook est de construire un **perceptron** From Scratch (en partant de zro). 

Le perceptron est le modle le plus de **rseau de neurones artificiels** car ne possdant qu'une **seule couche**. Il n'a qu'une seule sortie (boolenne)  laquelle toutes les entres (inputs) sont connectes (en appliquant une somme pondre).

La fonction d'activation (ou de transfert) est une fonction continue qui transforme la somme pondre des inputs. A partir de ce rsultat, on en dduit la sortie (output) du rseau. 

On peut citer comme fonctions d'activation la **unit step function (ou Heaviside)** et la **sigmoid function**.

### 2. Etapes de construction du perceptron

1. Coder la fonction **predict(W,X)** qui calcule $W.X +b$ (b dsigne le biais) pour chaque observation et renvoie 1 si le signe est positif 0 sinon

```python
def predict(W,row):
        activation = W[0]
        for i in range(len(row)-1):
            activation += W[i+1] * row[i]
        if activation > 0:
            return 1
        else:
            return 0
```

```python
for row in X:
    predictions = predict(W,row)
    print(predictions)
```

2. Coder la fonction **fit(X, Y, n_iterations,learning_rate)** qui met  jour W et b  chaque itration (formule dans l'algo ci-dessus)
A l'tape initiale, b = 0 et W = np.zeros(X.shape[1])
b est associ  la variable constante (valeurs gales  1)

```python
def fit(X,y,n_iterations,learning_rate):
    W = np.zeros(X.shape[1]+1)
    for i in range(n_iterations):
        sum_error = 0
        for count, row in enumerate(X):
            prediction = predict(W, row)
            error = y[count] - prediction
            sum_error += error**2
            W[0] = W[0] + learning_rate * error
            for z in range(len(row)):
                W[z+1] = W[z+1] + learning_rate * error * row[z]
        print('n_iteration : %d, learning_rate = %.3f, error = %.3f' % (i, learning_rate, sum_error))
    return W
    
```

```python
fit(X,y, 500, 0.01)
```

3.**Bonus** Crer un objet **Perceptron** qui intgre les fonctions ci-dessus 

```python
class Perceptron_raw:
  
  #constructor
    def __init__(self):
        self.w = None
        self.b = None
    
  #model  
    def model(self, x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0
  
  #predictor to predict on the data based on w
    def predict(self, X):
        result_list = []
        for x in X:
            result = self.model(x)
            result_list.append(result)
        return np.array(result_list)
    
    def fit(self, X, Y, epochs = 1, lr = 1):
        self.w = np.ones(X.shape[1])
        self.b = 0
        accuracy = {}
        max_accuracy = 0
        wt_matrix = []
    #for all epochs
        for i in range(epochs):
            for x, y in zip(X, Y):
                y_pred = self.model(x)
                if y == 1 and y_pred == 0:
                    self.w = self.w + lr * x
                    self.b = self.b - lr * 1
                elif y == 0 and y_pred == 1:
                    self.w = self.w - lr * x
                    self.b = self.b + lr * 1

            wt_matrix.append(self.w)    
            accuracy[i] = accuracy_score(self.predict(X), Y)
            if (accuracy[i] > max_accuracy):
                max_accuracy = accuracy[i]
                new_w = self.w
                new_b = self.b
    #checkpoint (Save the weights and b value)
        self.w = new_w
        self.b = new_b
        accuracy_list = [x for x in accuracy.values()]
        epoch_list = [x for x in range(epochs)]

        print(max_accuracy)
#         print(self.b)
    #plot the accuracy values over epochs
        plt.plot(epoch_list, accuracy_list)
        plt.xlabel("Epoch #")
        plt.ylabel("Accuracy")
        plt.ylim([0, 1])
        plt.show()
    
    #return the weight matrix, that contains weights over all epochs
        return np.array(wt_matrix), self.b
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# def plot_all()

```python
def plot_all(X, y, droite):
    # scatter plot, dots colored by class value
    df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
    colors = {0:'red', 1:'blue'}
    fig, ax = plt.subplots()
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])

 


    # Here i am calculating slope and intercept with given three weights
    for i in np.linspace(np.amin(X[:,:1]),np.amax(X[:,:1])):
        slope = -(droite[0]/droite[2])/(droite[0]/droite[1])  
        intercept = -droite[0]/droite[2]

 

        #y =mx+c, m is slope and c is intercept
        y = (slope*i) + intercept
        plt.plot(i, y, 'k.')

 

    plt.show()
```
