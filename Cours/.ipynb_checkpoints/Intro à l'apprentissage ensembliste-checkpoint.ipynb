#  Introduction  l'apprentissage ensembliste

On va introduire l'apprentissage ensembliste (**ensemble learning**) dont le but est d'amliorer les performances de notre modle en combinant plusieurs classifieurs dits faibles (en termes de **biais** ou **variance** des erreurs de prdiction).

<div>

</div>



Dans l'apprentissage ensembliste, on distingue 3 grandes familles de mthodes : le **Bagging** (parallle), le **Boosting** et le **Stacking** (squentielles)

Le **Bagging** qui est un diminutif de **B**ootstrap **Agg**regat**ing** (vous allez comprendre ce nom dans un instant) est une mthode permettant de combiner plusieurs classifieurs sur diffrents chantillons Bootstrap indpendants (tirage avec remise dans la base de dpart). La construction des classifieurs est faite de manire parallle. Cette mthode permet de rduire la variance des estimateurs individuels et offre une prdiction plus performante et plus stable, dpendante du nombre de classifieurs faibles utiliss. 

<div>

</div>

Le **Boosting** consiste  combiner plusieurs classifieurs en accordant plus de poids  ceux dont la prdiction est correcte. Le Boosting procde squentiellement. Chaque classifieur est une version adaptative du prcdent en donnant plus de poids aux observations mal ajustes ou mal prdites. Lestimateur construit  ltape k concentrera donc ses efforts sur les observations mal ajustes par lestimateur  ltape k1.Cette mthode permet de rduire le biais des estimateurs individuels.**AdaBoost** est l'algorithme de Boosting le plus populaire.

Le **stacking** est un procd qui consiste  appliquer un algorithme de machine learning  des classifieur gnrs par un autre algorithme de machine learning.Il sagit de prdire quels sont les meilleurs classifieurs et de les pondrer. 

<div>

</div>

On peut galement citer les **Random Forests** comme autre mthode ensembliste.Ils sont une **amlioration du Bagging** puisqu'elles crent des arbres plus diffrents structurellement en chantillonnant alatoirement seulement un sous-ensemble des features disponibles, ce qui permet de rduire la corrlation entre les arbres et donc d'obtenir un meilleur modle final.

</div>

L'agrgation des modles ensembliste se fait par **vote majoritaire** parmi les prdictions pour les problmes de classification et par **moyennisation** (pondre ou non) des prdictions pour des problmes de rgression.
