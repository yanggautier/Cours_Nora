# Les Matrices denses et creuses

### 1.Définition

#### Qu'est-ce qu'une matrice creuse ?

Dans la discipline de l'analyse numérique des mathématiques, une matrice creuse est une matrice contenant beaucoup de zéros.

Conceptuellement, les matrices creuses correspondent aux systèmes qui sont peu couplés. Si on considère une ligne de balles dont chacune est reliée à ses voisines directes par des élastiques, ce système serait représenté par une matrice creuse. Au contraire, si chaque balle de la ligne est reliée à toutes les autres balles, ce système serait représenté par une matrice dense. Ce concept de matrice creuse est très utilisé en analyse combinatoire et ses domaines d'applications tels que la théorie des réseaux, qui ont une faible densité de connexions.

Des matrices creuses de taille importante apparaissent souvent en science ou en ingénierie pour la résolution des équations aux dérivées partielles.

Quand on veut manipuler ou stocker des matrices creuses à l'aide de l'outil informatique, il est avantageux voire souvent nécessaire d'utiliser des algorithmes et des structures de données qui prennent en compte la structure peu dense de la matrice. Les opérations utilisant les structures et les algorithmes de base sur les matrices sont lents et utilisent une grande quantité de mémoire quand ils sont utilisés sur de grandes matrices creuses. Ces données sont facilement compressibles, et cette compression amène presque à chaque fois une baisse significative de la consommation mémoire. De plus, certaines matrices creuses de très grande taille ne sont pas manipulables par les algorithmes classiques.

##### Exemple

Considérons une matrice de taille 100 X 100 contenant seulement 10 éléments non nuls. Dans cette matrice, seuls 10 espaces sont remplis de valeurs non nulles et les espaces restants de la matrice sont remplis de zéro. Cela signifie que nous allouons totalement 100 X 100 X 2 = 20000 octets d'espace pour stocker cette matrice entière. Et pour accéder à ces 10 éléments non nuls, nous devons effectuer une numérisation 10000 fois. Pour simplifier, nous utilisons la représentation matricielle creuse suivante.

## 3.Problèmes de rareté
Des matrices clairsemées peuvent poser des problèmes de complexité d'espace et de temps. 

### 3.1 Complexité spatiale
En pratique, la plupart des grandes matrices sont rares.
Les très grandes matrices nécessitent beaucoup de mémoire. 

Excemple : Un exemple d'une très grande matrice qui est trop grande pour être stockée en mémoire est une matrice de liens 
qui montre les liens d'un site Web à un autre.
 
Il s'agit clairement d'un gaspillage de ressources mémoire car les valeurs nulles ne contiennent aucune information.

### 3.2 Complexité temporelle

Si la matrice contient principalement des valeurs nulles, c'est-à-dire aucune donnée, alors effectuer des opérations sur cette matrice peut prendre un certain temps où la majeure partie du calcul effectué impliquera l'ajout ou la multiplication de valeurs nulles ensemble.

## 4.Utilisation de matrices clairsemées


Scipy dispose de l' algorithme Compressed Sparse Row (CSR) qui convertit une matrice dense en matrice clairsemée, ce qui nous permet de compresser considérablement nos  données. 

Dans l'exemple ci-dessous, nous définissons une matrice clairsemée 3 x 6 comme un tableau dense.

```python
# dense to sparse
from numpy import array
from scipy.sparse import csr_matrix
# create dense matrix
A = array([[1, 0, 0, 1, 0, 0], [0, 0, 2, 0, 0, 1], [0, 0, 0, 2, 0, 0]])
#print(A)
# convert to sparse matrix (CSR method)
S = csr_matrix(A)
print(S)

```

```python
# reconstruct dense matrix
B = S.todense()
print(B)
```

NumPy ne fournit pas de fonction pour calculer la rareté d'une matrice.

Néanmoins, nous pouvons le calculer facilement en trouvant d'abord la densité de la matrice et en la soustrayant d'une. 
Le nombre d'éléments non nuls dans un tableau NumPy peut être donné par la fonction count_nonzero () et le nombre total 
d'éléments dans le tableau peut être donné par la propriété size du tableau. 

```python
sparsity = 1.0 - count_nonzero(A) / A.size
```

```python
#L'exemple ci-dessous montre comment calculer la rareté d'un tableau.
# calculate sparsity
from numpy import array
from numpy import count_nonzero
# create dense matrix
A = array([[1, 0, 0, 1, 0, 0], [0, 0, 2, 0, 0, 1], [0, 0, 0, 2, 0, 0]])
print(A)
# calculate sparsity
sparsity = 1.0 - count_nonzero(A) / A.size
print(sparsity)
```

 les matrices clairsemées ont beaucoup de valeurs nulles, nous pouvons appliquer des algorithmes spéciaux qui feront deux choses importantes:

1. compresser l'empreinte mémoire de notre objet matrice
2. accélérer de nombreuses routines d'apprentissage automatique

Étant donné que le stockage de toutes ces valeurs nulles est un gaspillage, nous pouvons appliquer des techniques de compression de données pour minimiser la quantité de données que nous devons stocker.
Les valeurs nulles peuvent être ignorées et seules les données ou les valeurs non nulles dans la matrice clairsemée doivent être stockées ou traitées.

```python
# Une matrice d'ensembles de données épars de 2 000 par 10 000 composés de zéros et de uns.

import numpy as np
np.random.seed(seed=12)  ## for reproducibility
dataset = np.random.binomial(1, 0.1, 20000000).reshape(2000,10000)  ## dummy data
y = np.random.binomial(1, 0.5, 2000)  ## dummy target variable
```

```python
# spy () de Matplotlib qui nous permet de visualiser la rareté d'un ensemble de données.

%matplotlib inline
import matplotlib.pyplot as plt
plt.spy(dataset)
plt.title("Sparse Matrix");
```

```python
# de comparaison, voici une matrice dense.
np.random.seed(seed=43)
plt.figure(figsize=(14,12))
plt.spy(np.random.binomial(1, 0.9, 20000000).reshape(2000,10000))
plt.title("Dense Matrix");
```

```python
#Nous avons une matrice de données appelée ensemble de données . Nous savons déjà qu'il est très rare, alors allons-y et 
#transformons-le avec la RSE de Scipy.
from scipy.sparse import csr_matrix
sparse_dataset = csr_matrix(dataset)
```

```python
# compresion 
import seaborn as sns

dense_size = np.array(dataset).nbytes/1e6
sparse_size = (sparse_dataset.data.nbytes + sparse_dataset.indptr.nbytes + sparse_dataset.indices.nbytes)/1e6

sns.barplot(['DENSE', 'SPARSE'], [dense_size, sparse_size])
plt.ylabel('MB')
plt.title('Compression')
```

Sur le graphique ci-dessus, nous pouvons voir que la matrice dense est de 160 Mo tandis que la matrice clairsemée 
est de 24 Mo. C'est une compression de 85%! Certes,nous avons commencé avec une matrice assez clairsemée.
