## 1. Qu'est-ce qu'un calcul matriciel?


### 1.1 Définition d'une matrice

Une matrice à n lignes et p colonnes sur un corps K est un tableau d'éléments
de K comportant n lignes et p colonnes.
On note aij l'élément d'une matrice A situé sur la ligne i et la colonne j. La matrice
A s'écrit :

 

On dit que A est  de taille (n,p)

Les nombres ($ a_{ij} $) sont appelés coefficients de la matrice A, et on note souvent A=($ a_{ij} $)1≤i≤n,1≤j≤p
L'ensemble des matrices à n lignes et p colonnes, à coefficients dans K, est noté Mn,p(K).

 - Si p = 1, A est une matrice colonne que l'on peut assimiler à un vecteur de Kn.
 - Si n = 1, A est une matrice ligne que l'on peut assimiler à une forme linéaire appartenant à (Kp)

Si n = p, A est une matrice carrée d'ordre n. Mn,n(K) se note Mn(K). 
Les éléments a11,. . . ,ann forment la diagonale principale de A.
Deux matrices A et B sont égales si elles sont de même format, et si ($ a_{ij} $) = ($ b_{ij} $) pour tout i ∈ {1,. . . ,n} et pour tout j ∈ {1,. . . ,p}.

#### Matrices particulières

Soit A = ($ a_{ij} $) une matrice carrée d'ordre n.

– A est triangulaire supérieure si $ a_{ij} $ = 0 pour i > j.




– A est triangulaire inférieure si ai j = 0 pour i < j



– A est diagonale si aij = 0 pour i différent de j





### 1.2 Matrice d'une application linéaire de E dans F

Soit E et F des espaces vectoriels de dimensions p et n, munis de bases respectives
B = (e1,. . . ,ep) et C = ( f1,. . . , fn) .
Soit f une application linéaire de E dans F. Elle est déterminée par la donnée des
vecteurs :




c'est-à-dire par la matrice A = $(a_{ij})$ dont les vecteurs colonnes sont les composantes
de f (ej ) dans la base de F qui a été choisie.
On dit que A est la matrice de f dans les bases B et C.

Y la matrice colonne des composantes de y dans C, M la matrice de f dans les bases
B et C.

```python
import numpy as np

#1. Création d'une matrice 
A = np.array([[15,-11,5],[20,-15,8],[8,-7,6]])
print(A)
print("la dimension de A est", A.shape)

#par concatenation selon l'axe vertical 
a1 = np.array([15,-11,5])
a2 = np.array([20,-15,8])
a3 = np.array([8,-7,6])
A = np.vstack((a1, a2,a3))
print(A)
print("la dimension de A est", A.shape)

# 2. Génération de matrices triangulaires

# Matrice triangulaire inférieur
Mat_triang_inf = np.tril(A)
print(Mat_triang_inf)

# Matrice triangulaire supérieur
Mat_triang_sup = np.triu(A)
print(Mat_triang_sup)

# 3. Matrice diagonale
# Extraction de la diagonale
d = np.diag(A)
print(d)
# Création dela matrice diagonale
D = np.diag(d)
print(D)

# 4. Matrice identité
I = np.identity(3)
print(I)

# 5. Matrice orthogonale
from numpy import array
from numpy.linalg import inv
# define orthogonal matrix
Q = np.array([
[1, 0],
[0, -1]])
print(Q)
# Calcul de l'inverse
V = np.linalg.inv(Q)
#Affichage de la transposée et de l'inverse
print(Q.T)
print(V)
print ("C'est la même matrice -> Q est orthogonale")

# Produit Q et sa transposee
I = (Q.T).dot(Q)
print(I)
```
